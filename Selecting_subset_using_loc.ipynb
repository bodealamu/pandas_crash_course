# Selecting a subset of a dataframe using Loc function

```python
import pandas as pd
```

```python
df = pd.read_csv('clean_auto_mpg.csv',
                 index_col='Unnamed: 0')
```

```python
df.sample()
```

```python
df.info()
```

## Select all rows and all columns

```python
df.loc[:,:]
```

## Select all rows and a few columns

```python
df.sample()
```

```python
df1 = df.loc[:,['car name','mpg']]
```

```python
df1.sample()
```

```python
df.loc[:,['car name','cylinders','displacement','horsepower']]
```
