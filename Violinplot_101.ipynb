```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('spotify.csv')
```

```python
df1 = df[df['track_genre'].isin(['acoustic', 
                                 'afrobeat', 
                                 'dancehall',
                                 'hip-hop',
                                 'jazz', 
                                 'r-n-b', 
                                 'reggae',])]
```

```python
df1.info()
```

```python
df1.nunique()
```

## Violin plots

Violin plots are very similar to Box plots in the way they function. They are both capable of plotting numeric continuous variables across discrete categories.

```python
plot = px.violin(data_frame=df1,
                x='track_genre',
                 y='speechiness',
                 title='Distribution of Speechiness across the genres'
                )
plot.show()
```

```python
plot = px.violin(data_frame=df1,
                x='track_genre',
                 y='danceability',
                 box=True,
                 title='Distribution of Speechiness across the genres'
                )
plot.show()
```

```python
plot = px.violin(data_frame=df1,
                x='track_genre',
                 y='danceability',
                 box=True,color='explicit',
                 title='Distribution of Speechiness across the genres'
                )
plot.show()
```

```python
plot = px.violin(data_frame=df1,
                 y='track_genre',
                 x='danceability',
                 box=True,
                 color='explicit',
                 title='Distribution of Danceability across the genres'
                )
plot.show()
```
