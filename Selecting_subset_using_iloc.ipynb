# Selecting a subset of a dataframe using iloc function

```python
import pandas as pd
```

```python
df = pd.read_csv('clean_auto_mpg.csv',
                 index_col='Unnamed: 0')
```

```python
df.sample()
```

```python
df.info()
```

## Select all rows and all columns

```python
df.iloc[:,:]
```

## Select all rows and a few columns

```python
df.sample()
```

```python
df1 = df.iloc[:,[0,1]]
```

```python
df1.head()
```

```python
df.iloc[:,[1,2,3,8]]
```

## Select all columns and a few rows

```python
df.sample()
```

```python
df.head()
```

```python
df.iloc[0:3,:]
```

```python
df.iloc[3:,:]
```

```python
df.iloc[:9,:]
```
