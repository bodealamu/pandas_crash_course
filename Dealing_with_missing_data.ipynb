## Determine the number of missing values present

```python
# How many missing values exist?
df.isna().sum()
```

## Deal with missing values by dropping columns or rows with missing values

```python
df.dropna(axis='columns') # drop any column with a missing value
```

```python
df.dropna(axis='index') # drop any row with a missing value
```

## Fill missing values with data

```python
df.Cabin.nunique()
```

```python
df['Age'].median()
```

```python
df['Embarked'].unique()
```

```python
missing_value_dict = {'Cabin':'missing',
                      'Embarked':'Unknown',
                      'Age':df['Age'].median()}
```

```python
df.fillna(missing_value_dict, inplace=True)
```

```python
# How many missing values exist?
df.isna().sum()
```
